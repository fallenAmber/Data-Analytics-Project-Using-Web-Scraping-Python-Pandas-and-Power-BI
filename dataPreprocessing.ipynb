```python
import pandas as pd

import json
```

```python
with open('json_files/wc_match_results.json') as f:
    
    data = json.load(f)
data[0]['matchSummary']
```

# Converting the json file to csv file using Pandas

```python
df = pd.DataFrame(data[0]['matchSummary'])
df.head()
```

```python
df.rename({'scorecard':'match_id'}, axis=1, inplace=True)
df.head()
```

```python
match_dict = {}

for index, row in df.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_dict[key1] = row['match_id']
    match_dict[key2] = row['match_id']
```

```python
match_dict
```

```python
df.to_csv('csv_files/match_summarycsv.csv', index = False)
```

# Processing batting summary files

```python
with open ('json_files/wc_batting_summary.json') as f:
    data_batting = json.load(f)
    
    record=[]
    
    for rec in data_batting:
        
        record.extend(rec["battingSummary"])
        
df_batting = pd.DataFrame(record)

df_batting.tail()
```

```python
df_batting["out/not out"] = df_batting.dismissal.apply(lambda x : "out" if len(x)>0 else "not out")
```

```python
df_batting
```

```python
df_batting['match_id'] = df_batting['match'].map(match_dict)
df_batting.head()
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
```

```python
df_batting.head(15)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head(15)
```

```python
df_batting.to_csv('csv_files/bating_summarycsv.csv', index = False)
```

# processing bowlling summary

```python
with open('json_files/wc_bowling_summary.json') as f:
    data_bowlling = json.load(f)
    record = []
    for rec in data_bowlling:
        record.extend(rec['bowlingSummary'])
record[:5]
```

```python
df_bowlling = pd.DataFrame(record)
print(df_bowlling.shape)
df_bowlling.head()
```

```python
df_bowlling['match_id'] = df_bowlling['match'].map(match_dict)
df_bowlling.head()
```

```python
df_bowlling.to_csv('csv_files/bowlling_summarycsv.csv', index = False)
```

# processing players information

```python
with open('json_files/wc_player_info.json') as f:
    data = json.load(f)
df_players = pd.DataFrame(data)
df_players.head(10)
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head()
```
